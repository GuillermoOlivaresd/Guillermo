```python
df= pd.read_csv('../input/datamic/mic.csv')
df.head()
```

```python
a_g= pd.pivot_table(index= ['Age','Sex'],
                          data = df,
                   aggfunc = len,
                   values = 'Availability')
a_g= a_g.rename(columns = {'Availability':'Count'},)
a_g = a_g.reset_index()
a_g
```

```python
fig1= sns.catplot(x='Age',
                 kind = 'count',
                 hue ='Sex',
                 data = df)
fig1
```

```python
fig2 = px.sunburst(a_g,
                    path=['Age','Sex'],
                   values = 'Count',
                  color_discrete_map = {'Sex': 'Orange'})


fig2.show()
```
