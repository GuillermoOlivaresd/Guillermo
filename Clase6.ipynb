```python
df= pd.read_excel('datasets/Canada.xlsx',
                  sheet_name = 'Canada by Citizenship',
                  skiprows= range(20),
                  skipfooter = 2,
                  engine ='openpyxl'
                 )
df.head(2)
```

```python
df.columns
```

```python
df.drop([ 'Type',    'Coverage',        'AREA',
           'REG',        'DEV',
           'DevName', 'Unnamed: 43',
       'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47',
       'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],
        axis = 1, inplace = True)
```

```python
df.columns
```

```python
df.rename(columns = {'OdName':'Country', 'AreaName':'Content', 'RegName':'Region'},
          inplace = True)
df.columns
```

```python
df.columns= list(map(str, df.columns))
```

```python
df.columns
```

```python
df['Total']=df.iloc[:, 3:].sum(axis=1)
df.head(2)
```

```python
df.iloc[:1, :]
```

```python
df.set_index('Country', inplace = True)
df.head(2)
```

```python
import matplotlib as mpl
mpl.style.use('seaborn')
years= list(map(str, range(1980, 2014)))
df.loc['Japan', years].plot()
```

```python
df.loc[['India','China'], years]
```

```python
ndf= df.loc[['China', 'Mexico', 'India','Afghanistan' ], years].transpose()
ndf.plot()
```

```python
import matplotlib.pyplot as plt
```

```python
ndf= df.loc[['China', 'Mexico', 'India','Afghanistan' ], years].transpose()
ndf.plot.area(alpha=0.5)
plt.text(25, 78000,'Max', fontsize = 10)
```
